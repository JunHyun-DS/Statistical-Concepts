<h4>White Noise & Random Walk are used to describe the stochastic and random movement of nature </h4>

Suppose, there's a problem of throwing dice and matching number.

We know that the probability of dice is $\frac{1}{6}$, but we can't say the number deterministically like 1 or 2.

Those results are random and non-deterministic. 

Above random and non-deterministic problems are called <b>Uncertainty of nature</b>.

This simple fliping dice and matching problem involves uncertainty, the more complex the problem is, the greater the uncertainty will be.

So, White Noise and Random Walk concept are used to describe the uncertainity of nature.

```r
# Parameter
t <- 0:1000 # time
sig2 <- 0.01 # variance of random walk
```

<h2> White Noise </h2>

White Noise is a random signal with $x_{t} $~$ N(0,Ïƒ^2)$

```r
options(repr.plot.width=8, repr.plot.height=3)
x <- rnorm(n = length(t) - 1, sd = sqrt(sig2))
plot(x, type = "l", xlab='Time')
```


<h2> Random Walk </h2>


A random walk is a timeseries model $x_{t}$ such that $x_{t} = x_{t-1}+w_{t}$, where $w_{t}$ is discrete white noise series.



<h3> $x_{t} = x_{t-1} + w_{t}$  </h3> 

$w_{t}$ : discrete white noise

$x_{t} = Bx_{t} + w_{t} = x_{t-1} + w_{t}$

$x_{t-1} = Bx_{t-1} + w_{t-1} = x_{t-2} + w_{t-1}$

$x_{t} = (1+B+B^2+...)w_{t}$ &rarr; $x_{t} = w_{t}+w_{t-1}+w_{t-2}+... $

```r
# Random Work
xcum <- c(0, cumsum(x))
plot(t, xcum, type = "l", xlab='Time')
```
